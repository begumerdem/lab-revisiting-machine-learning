# Train Test split for the model

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
features = num_cat
labels = target
display(features.head())
display(labels.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20, random_state = 1)
```

```python
X_train
```

# One-hot Encoding the categorical columns

```python
X_train_hot_col = X_train[['STATE', 'HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A']]
X_train_hot_col
```

```python
X_test_hot_col = X_test[['STATE', 'HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A']]
X_test_hot_col
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_hot_col)

cols = encoder.get_feature_names(input_features = X_train_hot_col.columns)

X_train_col_encoded = pd.DataFrame(encoder.transform(X_train_hot_col).toarray(),columns=cols)

X_train_col_encoded
```

```python
cols1 = encoder.get_feature_names(input_features = X_test_hot_col.columns)

X_test_col_encoded = pd.DataFrame(encoder.transform( X_test_hot_col).toarray(),columns=cols1)

X_test_col_encoded
```

```python
X_train = X_train.drop(['STATE', 'HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A'],axis = 1)
X_train=X_train.reset_index(drop=True)
```

```python
X_train 
```

```python
X_test = X_test.drop(['STATE', 'HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A'],axis = 1)
X_test=X_test.reset_index(drop=True)
```

```python
X_test
```

```python
X_train_encoded= pd.concat([X_train,X_train_col_encoded],axis=1)
X_train_encoded= X_train_encoded.reset_index()
X_train_encoded= X_train_encoded.drop('index',axis=1)
```

```python
X_test_encoded= pd.concat([X_test,X_test_col_encoded],axis=1)
X_test_encoded= X_test_encoded.reset_index()
X_test_encoded= X_test_encoded.drop('index',axis=1)
X_test_encoded
```

```python
for colname, coltype in X_train_encoded.dtypes.iteritems():
    print(colname, coltype)
    
    
# checked if I missed any categorical columns, looking at X_train all data types
```

# Min-Max scaling for the numerical columns

```python
transformer = MinMaxScaler().fit(X_train_encoded)
X_norm1 = transformer.transform(X_train_encoded)
X_train_scaled = pd.DataFrame(X_norm1, columns=X_train_encoded.columns)
X_train_scaled = X_train_scaled.reset_index(drop=True)
```

```python
X_norm2 = transformer.transform(X_test_encoded)
X_test_scaled = pd.DataFrame(X_norm2, columns=X_test_encoded.columns)
X_test_scaled= X_test_scaled.reset_index(drop=True)
```

```python
X_train_scaled
```

# Build a model

```python
X_train = X_train_scaled
X_test = X_test_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
print("test data accuracy was: ",clf.score(X_test, y_test))
print("train data accuracy was: ",clf.score(X_train, y_train))
```
